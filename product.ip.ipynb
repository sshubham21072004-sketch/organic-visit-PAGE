```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('data.xlsx') #load excel file
```

```python
df
```

CHECK DUPLICATES

```python
df.duplicated() 
```

```python
df = df.drop_duplicates()
```

CHECK MISSING VALUES

```python
df.isnull().sum()
```

REMOVE UNWANTED COLUMNS

```python
df.drop(columns=['Unnamed col'], inplace=True)
```

RENAME COLUMNS

```python
df.rename(columns={
    'Pid': 'Product_ID',
    'Product Name': 'Product',
    'Entry-date': 'Entry_Date'
}, inplace=True)
```

MAINTAIN PRICE IN RUPEES ONLY 
REMOVE $ AND CONVERT TO NUMRIC

```python
df['Cost'] = df['Cost'].astype(str).str.replace('$','')
df['Cost'] = pd.to_numeric(df['Cost'], errors='coerce')
```

Split Model column using separators

Model examples:

L1-1997

H3:2022

```python
df[['Model_Code', 'Model_Year']] = df['Model'].str.split('[-:]', expand=True)
```

Extract month from Entry_Date into new column

```python
df['Entry_Date'] = pd.to_datetime(df['Entry_Date'], errors='coerce')

df['Month'] = df['Entry_Date'].dt.month_name()
```

Create new DataFrame with

Product, Cost, Amazon, Flipkart (3 columns)
(using melt)

```python
new_df = df.melt(
    id_vars=['Product', 'Cost'],
    value_vars=['Amazon', 'Flipkart'],
    var_name='Platform',
    value_name='Available'
)
```

Keep only available products

```python
new_df = new_df[new_df['Available'] == 1]
```
